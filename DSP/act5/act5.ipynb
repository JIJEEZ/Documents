```python
import numpy as np
import matplotlib.pyplot as plt
```

y = sin(2 * pi * frequency * time)

```python
t1 = np.arange(0,1,0.001)
f1 = 1
y1 = np.sin(2*np.pi*f1*t1)
```

```python
t2 = np.arange(0,1,0.001)
f2 = 5
y2 = np.sin(2*np.pi*f2*t2)
```

```python
t3 = np.arange(0,1,0.001)
f3 = 10
y3 = np.sin(2*np.pi*f3*t3)
```

```python
plt.figure(figsize=(10,5))
plt.plot(t1,y1,color='r')
plt.plot(t2,y2,color='g')
plt.plot(t3,y3,color='b')
plt.grid(True)
plt.xlabel('Time (s)',position=(0.95,1))
plt.ylabel('Time (v)',position=(1,0.09))
ax = plt.gca()

plt.show()
```
